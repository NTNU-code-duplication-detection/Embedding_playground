```python
java_code = """
public class L1 {

    public static void main(String[] args) {
        // Variables
        int sum = 0;
        int product = 1;
        int count = 0;
        int largest = 0;
        int smallest = 100;

        // For loop: calculate sum of 1 to 20
        System.out.println("=== For Loop: Sum of 1 to 20 ===");
        for (int i = 1; i <= 20; i++) {
            sum += i;
            if (i % 5 == 0) {
                System.out.println("Running sum at " + i + ": " + sum);
            }
        }
        System.out.println("Total sum: " + sum);

        // For loop: find even and odd numbers
        System.out.println("\n=== For Loop: Even and Odd Numbers ===");
        for (int i = 1; i <= 15; i++) {
            if (i % 2 == 0) {
                System.out.println(i + " is even");
            } else {
                System.out.println(i + " is odd");
            }
        }

        // For loop: multiplication table
        System.out.println("\n=== For Loop: Multiplication Table of 7 ===");
        for (int i = 1; i <= 10; i++) {
            int result = 7 * i;
            if (result > 50) {
                System.out.println("7 x " + i + " = " + result + " (above 50!)");
            } else {
                System.out.println("7 x " + i + " = " + result);
            }
        }

        // While loop: countdown
        System.out.println("\n=== While Loop: Countdown ===");
        int countdown = 10;
        while (countdown > 0) {
            if (countdown <= 3) {
                System.out.println(countdown + " ... almost there!");
            } else {
                System.out.println(countdown);
            }
            countdown--;
        }
        System.out.println("Liftoff!");

        // While loop: find largest and smallest in a set
        System.out.println("\n=== While Loop: Find Largest and Smallest ===");
        int[] numbers = {12, 45, 7, 89, 23, 56, 3, 67, 34, 91};
        int index = 0;
        while (index < numbers.length) {
            if (numbers[index] > largest) {
                largest = numbers[index];
            }
            if (numbers[index] < smallest) {
                smallest = numbers[index];
            }
            index++;
        }
        System.out.println("Largest: " + largest);
        System.out.println("Smallest: " + smallest);

        // While loop: factorial calculator
        System.out.println("\n=== While Loop: Factorial ===");
        int num = 8;
        int factNum = num;
        long factorial = 1;
        while (num > 0) {
            factorial *= num;
            num--;
        }
        System.out.println("Factorial of " + factNum + " is " + factorial);

        // For loop with nested if: grade checker
        System.out.println("\n=== For Loop: Grade Checker ===");
        int[] scores = {95, 82, 67, 45, 73, 88, 54, 91, 60, 78};
        for (int i = 0; i < scores.length; i++) {
            if (scores[i] >= 90) {
                System.out.println("Score " + scores[i] + ": Grade A");
            } else if (scores[i] >= 80) {
                System.out.println("Score " + scores[i] + ": Grade B");
            } else if (scores[i] >= 70) {
                System.out.println("Score " + scores[i] + ": Grade C");
            } else if (scores[i] >= 60) {
                System.out.println("Score " + scores[i] + ": Grade D");
            } else {
                System.out.println("Score " + scores[i] + ": Grade F");
            }
        }

        // While loop: sum until threshold
        System.out.println("\n=== While Loop: Sum Until Threshold ===");
        int total = 0;
        int step = 1;
        while (total < 100) {
            total += step;
            if (total > 50 && count == 0) {
                System.out.println("Passed 50 at step " + step);
                count = 1;
            }
            step++;
        }
        System.out.println("Reached " + total + " at step " + (step - 1));
    }
}
"""


print(java_code)
```
